```python
# Import of needed libraries for data pre-processing
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import pos_tag
from nltk import bigrams
from nltk import trigrams

# Stopword definition in terms of taking the English stopwords
stopwords = stopwords.words('english')

# Tokenization to split job titles into words
# w.lower() converts all words into lowercase to make sure capitalization is not leading to considering the same word differently 
# stopwords are removed to filter out low-value information

def tokenize (column):
    column = re.sub(r'/', ' ', column)
    token_data = nltk.word_tokenize(column)
    
    clean_tokens = [re.sub(r'[^a-zA-Z]', '', w).lower() for w in token_data]
    return [w for w in clean_tokens if w and w not in stopwords] 

df['token_data'] = df['job_title'].apply(lambda x: tokenize(x))

# Lemmatization with WordNetLemmatizer thereby only including nouns and verbs to reduce the noise
lm = WordNetLemmatizer()

def pos_tagging (token_data):
    pos_tags = pos_tag(token_data)
    filtered_tokens = [word for word, pos in pos_tags if pos.startswith('N') or pos.startswith('V')]
    return filtered_tokens

df['filtered_data']=df['token_data'].apply(pos_tagging)

df['lemmatized_tokens'] = df['filtered_data'].apply(lambda token_list: [lm.lemmatize(token) for token in token_list])


# Bigram creation to identify which two words are occurring most often as a basis for identifying most common job titles
def create_bigrams(lemmatized_tokens):
    bi_grams = list(bigrams(lemmatized_tokens))
    return [' '.join(pair) for pair in bi_grams]

df['bigrams']=df['lemmatized_tokens'].apply(create_bigrams)


# Analogous to bigrams: creation of trigrams to identify which three words are occurring most often as a basis for identifying most common job titles
def create_trigrams(lemmatized_tokens):
    tri_grams = list(trigrams(lemmatized_tokens))
    return [' '.join(triple) for triple in tri_grams]

df['trigrams']=df['lemmatized_tokens'].apply(create_trigrams)

#pd.set_option('display.max_rows', None)

df[['job_title','token_data','filtered_data','lemmatized_tokens','bigrams','trigrams']]
```

```python
# Predefined job title categories
# The first time of running the code, the following job_title_categories definition was used: 
#job_title_categories = {
#    "Machine Learning Scientist": "machine learning",
#    "Data Engineer": "engineer",
#    "Data Analyst": "analyst",
#    "Data Scientist": "scientist" 
#}

# Redefinition of the job_title_categories definition to also include additional terms to make sure all job_title have a respective job_title_category
job_title_categories = {
    "Machine Learning Scientist": "machine learning",
    "Data Engineer": ["engineer","modeler","architect"],
    "Data Analyst": ["analyst","analytics"],
    "Data Scientist": ["scientist","science"]
}

# Function to categorize job titles
# First, the column "bigrams" is checked
# Given that some job_title do not result in a bigram as the job_title consists of less than two words after filtering out everyting <> nouns and <> verbs, "job_title" is checked in a second instance
def categorize_job_title(row):
    bigrams = row['bigrams']
    job_title = row['job_title'].lower()
    
    words = set()
    for bigram in bigrams:
        individual_words = bigram.split()
        words.update(individual_words)
        words.add(bigram)
    
    words.update(job_title.split())
    
    for category, keywords in job_title_categories.items():
        if isinstance(keywords, str):
            if keywords in words:
                return category
        else:
            if any(keyword in words for keyword in keywords):
                return category
    return pd.NA

# Apply the function to create the new column "job_title_category"
df['job_title_category'] = df.apply(categorize_job_title, axis=1)

# If no category matches, assign NA
df['job_title_category'].fillna(value=pd.NA, inplace=True)

# Show the top 20 results
df[['job_title','bigrams','job_title_category']].head(20)
```

```python
# Creation of "job_senior" column
seniority_categories = {
    "yes": ["principal","senior", "sr", "experienced", "iii","expert"]}

# Function to assign "yes"
# Only the column "job_title" is checked as respective information has partially been filtered out during pos tagging
def assign_seniority(row):
    job_title = row['job_title'].lower()
    
    for category, keywords in seniority_categories.items():
        if any(keyword in job_title for keyword in keywords):
            return category
    return "no"

# Apply the function to create the new column "job_senior"
df['job_senior'] = df.apply(assign_seniority, axis=1)

# Show the top 5 results with job_senior = "yes"
df_senior_yes_rows = df[df['job_senior']=="yes"]

df_senior_yes_rows[['job_title','job_title_category','job_senior']].head()
```

```python
# Creation of "job_lead" column
lead_categories = {
    "yes": ["president","lead", "director", "vp"]
}

# Function to assign "yes"
# Only the column "job_title" is checked as respective information has partially been filtered out during pos tagging
def assign_lead(row):
    job_title = row['job_title'].lower()
    
    for category, keywords in lead_categories.items():
        if any(keyword in job_title for keyword in keywords):
            return category
    return "no"

# Apply the function to create the new column "job_lead"
df['job_lead'] = df.apply(assign_lead, axis=1)

# Show the top 5 results with job_lead = "yes"
df_lead_yes_rows = df[df['job_lead']=="yes"]

df_lead_yes_rows[['job_title','job_title_category','job_lead']].head()
```

```python
# Get all locations which do not follow the typical structure of location --> comma --> two-letter code for the state
pattern = r', [A-Z]{2}$'
filtered_df = df[~df['location'].str.contains(pattern)]

print(filtered_df['location'])
```

```python
# Dictionary to map full state names to their abbreviations
state_abbreviations = {
    'California': 'CA','New Jersey': 'NJ','Texas': 'TX', 'Utah': 'UT'}

# Drop rows where "location" = "United States" as no location information can be retrieved
df = df[df['location'] != 'United States']

# Function to process the location and extract city and state
def process_location(location):
    # Check if the location follows the comma and two-letter pattern
    if re.search(r', [A-Z]{2}$', location):
        city = location.rsplit(', ', 1)[0]
        state = location[-2:]
        return city, state
    # Handle special case "Remote"
    elif location in ["Remote"]:
        return pd.NA, pd.NA
    # Check if the location contains a full state name and, if yes, keep city empty and fill state with the abbreviation
    else:
        for full_state, abbr in state_abbreviations.items():
            if location.endswith(full_state):
                city = location.rsplit(', ', 1)[0] if ', ' in location else pd.NA
                return city, abbr
    return "", ""

# Apply the function to create new columns
df[['location_city_us', 'location_state_us']] = df['location'].apply(
    lambda loc: pd.Series(process_location(loc))
)

# Creation of the "location_remote" column
df['location_remote'] = df['location'].apply(lambda loc: 'yes' if loc == 'Remote' else 'no')

df[['location','location_city_us','location_state_us','location_remote']].head()
```

```python
# Three new columns "headquarter_city", "headquarter_state" and "headquarter_country" are created
# For the rows following the typical structure:
# (a) "headquarter_city" will be filled with everything before the comma
# (b) "headquarter_state" will be filled with the last two letters. To show the full state name as for the non-US states (cf. below), the state name will be matched from the csv file "uscities"
# (c) "headquarter_country" will be filled with "US"
# For the rows not following the typical structure:
# (a) "headquarter_city" will be filled with everything before the comma
# (b) "headquarter_state" will be filled with the column "admin_name" from the csv file "worldcities"
# (c) "headquarter_country" will be filled with the column "iso2" from the csv file "worldcities"

import numpy as np

# Pattern to match strings ending with ', XX' where XX are uppercase letters
pattern = r', [A-Z]{2}$'

# Extract the city part (everything before the comma)
df['headquarter_city'] = df['headquarters'].str.split(',').str[0]

# Check which entries match the pattern
matches_pattern = df['headquarters'].str.contains(pattern)

# Extract the state for the matching entries
df['headquarter_state_prework'] = np.where(matches_pattern, df['headquarters'].str[-2:], np.nan)

# Assign 'US' to headquarter_country if state is present, otherwise extract the country part (everything after the comma)
df['headquarter_country_prework'] = np.where(
    matches_pattern, 
    'US', 
    df['headquarters'].str.split(',').str[1].str.strip()
)

df[['headquarters','headquarter_city','headquarter_state_prework','headquarter_country_prework']].head(50)
```

```python
# Get the externally retrieved .csv file "worldcities.csv"
worldcities = pd.read_csv('worldcities.csv')

# Create a dictionary to map iso2 codes to country names
iso2_to_country = dict(zip(worldcities['country'], worldcities['iso2']))

# Create a new column for headquarter_country_2
df['headquarter_country'] = df['headquarter_country_prework']
df.loc[~matches_pattern, 'headquarter_country'] = df.loc[~matches_pattern, 'headquarter_country_prework'].map(iso2_to_country)

df[['headquarters','headquarter_city','headquarter_state_prework','headquarter_country_prework','headquarter_country']].head(80)
```
